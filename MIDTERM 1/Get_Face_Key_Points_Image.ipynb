```python
def feature_centroid_dx(kps: dict) -> float:
    (cx, cy) = kps['C']
    (nx, ny) = kps['N']
    (mx, my) = kps['M']
    D = kps['D'] if kps['D'] != 0 else 1e-6

    gx = (cx + nx + mx) / 3.0
    dx = (gx - cx) / D
    return dx

dx = feature_centroid_dx(kps)
dx
```

```python
def feature_nose_offset(kps: dict) -> float:
    (cx, cy) = kps['C']
    (nx, ny) = kps['N']
    D = kps['D'] if kps['D'] != 0 else 1e-6
    return (nx - cx) / D

s1 = feature_nose_offset(kps)
s1
```

```python
def feature_mouth_offset(kps: dict) -> float:
    (cx, cy) = kps['C']
    (mx, my) = kps['M']
    D = kps['D'] if kps['D'] != 0 else 1e-6
    return (mx - cx) / D

s4 = feature_mouth_offset(kps)
s4
```
